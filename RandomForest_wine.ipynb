```python
from sklearn.ensemble  import RandomForestClassifier 
from sklearn.model_selection import train_test_split
import pandas as pd

df=pd.read_csv("./ML_data/wine.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
x=df.iloc[:,1:14]
y=df.iloc[:,0]
x_train, x_test, y_train, y_test=train_test_split(x,y ,test_size=0.2)
random_forest_clf=RandomForestClassifier(n_estimators=10, min_samples_split=30)
random_forest_clf.fit(x_train,y_train)
score=random_forest_clf.score(x_test, y_test)
print("",score)

```
